```python
import pandas as pd
```

```python
df=pd.read_csv('../../Datasets/IMDB Dataset.csv')
df.shape
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data=train_test_split(df,test_size=0.2,random_state=11)
```

```python
train_data.shape
```

```python
reviews=train_data['review']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer(num_words=20000) #we took most frequent 20k words
tokenizer.fit_on_texts(reviews) #it only finds and fits distinct words from reviews
```

```python
reviews_tokenized=tokenizer.texts_to_sequences(reviews) #convert text to sequence
```

```python
len(reviews_tokenized)
```

```python
print(reviews_tokenized[0]) #tokenised review no. 1
```

```python
reviews.iloc[0] #review number 1
```

```python
len(reviews_tokenized[0])
```

```python
tokenizer.index_word 
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
reviews_processed=pad_sequences(reviews_tokenized,maxlen=180)
```

```python
len(reviews_processed[0]),len(reviews_processed[1]),
```

```python
(reviews_processed[0]) 
```

```python
from tensorflow.keras.layers import Embedding,LSTM,Dense
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Embedding(20000,128)) #length of reviews=20000 convert to 128=length of wordvectors
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',metrics='accuracy')
```

```python
y=train_data['sentiment'].map({'positive':1,'negative':0})
```
